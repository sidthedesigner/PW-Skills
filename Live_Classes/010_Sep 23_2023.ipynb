# 2. Time Module

In Python, you can use the `time` module to add delays or pauses to your program, which can simulate sleep or waiting. The most common function for introducing a sleep or delay in Python is `time.sleep()`.

Here's how you can use `time.sleep()`:

```python
import time

# Sleep for 2 seconds
time.sleep(2)
```

In this example, the program will pause execution for 2 seconds before continuing. You can replace `2` with the number of seconds you want to sleep for.

Here's a more detailed example of how to use `time.sleep()` in a program:

```python
import time

def main():
    print("Start of the program")
    
    # Sleep for 3 seconds
    time.sleep(3)
    
    print("After 3 seconds")
    
    # Sleep for 2 seconds
    time.sleep(2)
    
    print("End of the program")

if __name__ == "__main__":
    main()
```

When you run this program, it will display "Start of the program," then pause for 3 seconds, display "After 3 seconds," pause for an additional 2 seconds, and finally display "End of the program."

You can use `time.sleep()` to introduce delays in your code for various purposes, such as controlling the rate of API requests, creating periodic tasks, or simulating real-time behavior.

````python
# Program for Vending Machine (Approach 1)

class VendingMachine:
    def __init__(self):
        self.items = {
            'Coke': 1.50,
            'Chips': 1.00,
            'Candy': 0.75,
            'Water': 1.25
        }
        self.balance = 0.0

    def display_items(self):
        print("Available Items:")
        for item, price in self.items.items():
            print(f"{item}: ${price:.2f}")

    def insert_money(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Inserted ${amount:.2f}. Your balance: ${self.balance:.2f}")
        else:
            print("Invalid amount. Please insert a positive amount.")

    def select_item(self, item_name):
        if item_name in self.items:
            item_price = self.items[item_name]
            if self.balance >= item_price:
                self.balance -= item_price
                print(f"You have purchased {item_name}. Remaining balance: ${self.balance:.2f}")
            else:
                print("Insufficient balance. Please insert more money.")
        else:
            print("Invalid item. Please select a valid item.")

    def return_change(self):
        change = self.balance
        self.balance = 0
        return change


def main():
    vending_machine = VendingMachine()

    while True:
        print("\nWelcome to the Vending Machine!")
        vending_machine.display_items()
        print(f"Your current balance: ${vending_machine.balance:.2f}")

        choice = input("\nEnter 'I' to insert money, 'S' to select an item, 'C' to return change, or 'Q' to quit: ").strip().upper()

        if choice == 'I':
            amount = float(input("Enter the amount you want to insert: "))
            vending_machine.insert_money(amount)
        elif choice == 'S':
            item_name = input("Enter the item name you want to purchase: ")
            vending_machine.select_item(item_name)
        elif choice == 'C':
            change = vending_machine.return_change()
            print(f"Change returned: ${change:.2f}")
        elif choice == 'Q':
            print("Thank you for using the Vending Machine!")
            break
        else:
            print("Invalid choice. Please select a valid option.")


if __name__ == "__main__":
    main()
````

This program defines a VendingMachine class that allows you to display available items, insert money, select items, and return change. The main function provides a simple command-line interface for users to interact with the vending machine.

# 3. Recursive Call

In Python, a recursive call refers to a situation where a function calls itself as part of its own execution. This is a fundamental concept in recursive programming and is often used to solve problems that can be broken down into smaller, similar subproblems. When a function calls itself, it does so with a modified set of arguments or state, typically moving closer to a base case or termination condition to eventually stop the recursion.

Here's a basic structure of a recursive function in Python:

```python
def recursive_function(parameters):
    # Base case: Check if the recursion should stop.
    if base_case_condition(parameters):
        return base_case_value

    # Recursive case: Modify parameters and make a recursive call.
    new_parameters = modify_parameters(parameters)
    result = recursive_function(new_parameters)

    # Optionally, process the result and return it.
    return process_result(result)
```

Key components of a recursive function:

1. **Base Case:** A condition that determines when the recursion should stop. When this condition is met, the function returns a specific value without making any further recursive calls. It prevents infinite recursion.

2. **Recursive Case:** This is where the function calls itself with modified parameters. The idea is to break down the problem into smaller, similar subproblems.

3. **Modifying Parameters:** Inside the recursive case, the function typically modifies the parameters or arguments passed to itself, ensuring progress toward the base case.

4. **Processing the Result:** After the recursive call returns, the result may be processed or combined with other results before being returned by the current invocation.

Here's a simple example of a recursive function that calculates the factorial of a number:

```python
def factorial(n):
    # Base case: factorial(0) is defined as 1.
    if n == 0:
        return 1
    # Recursive case: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)
```

In this example, `factorial` calls itself with a smaller value `n - 1` until it reaches the base case (`n == 0`), at which point it stops and returns the result `1`.

````python
# Program to calculate the sum of digits of a number using recursion

def sum_of_digits_recursive(n):
    # Base case: if the number is a single digit, return the number itself
    if n < 10:
        return n
    else:
        # Recursive case: add the last digit to the sum of digits of the rest of the number
        last_digit = n % 10
        rest_of_number = n // 10
        return last_digit + sum_of_digits_recursive(rest_of_number)

# Input number
num = int(input("Enter a number: "))

# Calculate and print the sum of digits
result = sum_of_digits_recursive(num)
print("Sum of digits:", result)
````

This code defines a recursive function sum_of_digits_recursive that calculates the sum of digits. It first checks if the number is a single digit (base case) and returns it. If the number has more than one digit, it extracts the last digit, calculates the sum of digits for the rest of the number, and adds them together.

**Note:**

In Python, `%` and `//` are operators used for performing arithmetic operations on numbers. Here's an explanation of both operators:

1. **`%` (Modulus Operator):**
   - The modulus operator `%` calculates the remainder when one number is divided by another.
   - For example, `n % 10` computes the remainder when the number `n` is divided by 10. This operation effectively extracts the last digit of the number `n`.
   - For instance, if `n` is 123, `n % 10` would result in 3 because 123 divided by 10 gives a quotient of 12 with a remainder of 3.

2. **`//` (Floor Division Operator):**
   - The floor division operator `//` performs division and rounds down to the nearest integer.
   - For example, `n // 10` calculates the integer division of `n` by 10. It effectively removes the last digit from the number.
   - If `n` is 123, `n // 10` would result in 12 because it's the integer division of 123 by 10.

Together, when you use `n % 10` and `n // 10` in combination, you can extract the last digit of a number `n` and obtain the remaining digits as well. This is useful when you want to process individual digits of a number, as demonstrated in the previous code examples for calculating the sum of digits.

````python
# Program to use recursion to find the maximum number in a list

def find_max_recursive(lst):
    # Base case: If the list has only one element, return that element
    if len(lst) == 1:
        return lst[0]
    
    # Recursive case: Compare the first element with the maximum of the rest of the list
    else:
        max_rest = find_max_recursive(lst[1:])
        if lst[0] > max_rest:
            return lst[0]
        else:
            return max_rest

# Example usage:
my_list = [3, 7, 2, 8, 5]
max_number = find_max_recursive(my_list)
print("The maximum number in the list is:", max_number)
````

Let's break down the `find_max_recursive` function step by step:

```python
def find_max_recursive(lst):
    # Base case: If the list has only one element, return that element
    if len(lst) == 1:
        return lst[0]
```

1. The function `find_max_recursive` takes a list `lst` as its input.

2. In the base case, it checks if the length of the list `lst` is equal to 1. If it is, this means there's only one element in the list. In this case, we immediately return that element as it must be the maximum.

```python
    else:
        max_rest = find_max_recursive(lst[1:])
        if lst[0] > max_rest:
            return lst[0]
        else:
            return max_rest
```

3. In the recursive case, if the list has more than one element, the function proceeds as follows:

   a. It calls itself recursively with the sublist `lst[1:]`. This recursive call finds the maximum number in the rest of the list, excluding the first element.

   b. The result of the recursive call is stored in the `max_rest` variable.

   c. It then compares the first element of the list, `lst[0]`, with the `max_rest` obtained from the recursive call.

   d. If `lst[0]` is greater than `max_rest`, it means that the maximum number is in the first position of the list, so it returns `lst[0]` as the maximum.

   e. If `max_rest` is greater, it means the maximum number is in the rest of the list, so it returns `max_rest` as the maximum.

This recursive process continues until we reach the base case (a list with only one element), and then it starts unwinding and returning the maximum value from each level of recursion until we get the maximum number from the original list.

Here's an example to illustrate how the function works:
```python
my_list = [3, 7, 2, 8, 5]

# Calling find_max_recursive(my_list)
# 3 > find_max_recursive([7, 2, 8, 5])
#   7 > find_max_recursive([2, 8, 5])
#     2 > find_max_recursive([8, 5])
#       8 > find_max_recursive([5])
#       Base case reached, return 5
#     Return max(2, 5) = 5
#   Return max(7, 5) = 7
# Return max(3, 7) = 7
# The maximum number in the list is 7
```

````python
# Program for calculating a number raised to a power using recursion

def power(base, exponent):
    # Base case: If the exponent is 0, return 1
    if exponent == 0:
        return 1
    # Recursive case: Calculate base^exponent-1 and multiply it by base
    else:
        return base * power(base, exponent - 1)

# Example usage:
base = 5
exponent = 3
result = power(base, exponent)
print(f"{base}^{exponent} = {result}")
````

# 4. Using Recursion in Production Code

Recursion is a powerful and elegant programming technique, but it's important to be cautious when using it in production code, including in Python, for several reasons:

1. **Stack Overflow**: Recursive functions rely on the call stack to keep track of function calls. If your recursion goes too deep, it can lead to a stack overflow error. Python has a limit on the maximum recursion depth (typically around 1000), so deep recursion can cause your program to crash.

2. **Performance Overhead**: Recursive function calls often come with a performance overhead compared to iterative solutions. Each function call requires memory allocation for its own stack frame, and this can add up quickly in terms of both memory usage and execution time.

3. **Readability and Maintainability**: Recursive code can be less readable and harder to maintain than equivalent iterative code. Understanding the flow of a recursive function can be challenging, especially for developers not familiar with the specific implementation.

4. **Limited Tail Call Optimization**: Python's standard CPython interpreter does not perform tail call optimization (TCO), which means that recursive functions may not be as memory-efficient as their iterative counterparts. Some other languages and interpreters do offer TCO.

5. **Debugging Complexity**: Debugging recursive code can be more challenging than debugging iterative code. Stepping through recursive calls in a debugger can be cumbersome, especially in complex recursive algorithms.

6. **Resource Consumption**: Recursive functions can consume a significant amount of memory due to the call stack, which may not be suitable for systems with limited memory resources.

Despite these considerations, there are situations where recursion is a natural and elegant solution to a problem, and in those cases, it should be used. For example, recursive algorithms are often a good choice for problems involving tree structures (e.g., binary trees) or when the problem itself is inherently recursive (e.g., calculating Fibonacci numbers).

However, in production code where performance, readability, and maintainability are critical, it's often advisable to favor iterative solutions or consider tailoring your recursive code to be more memory-efficient. Additionally, you can use Python's built-in limit on recursion depth as a guideline and consider alternative approaches when dealing with deep recursion.
